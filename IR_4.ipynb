```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering

# Generate synthetic two-dimensional data
X, y = make_blobs(n_samples=150, centers=3, cluster_std=0.60, random_state=0)

# Plotting the generated data
plt.figure(figsize=(10, 7))
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='o', s=50)
plt.grid()
plt.show()
```

```python
# Now we will apply Agglomerative Clustering to this dataset

# Create the model with 3 clusters
agg_cluster_model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')

# Fit model to the data
y_agg = agg_cluster_model.fit_predict(X)

# Visualize the clusters
plt.figure(figsize=(10, 7))
plt.scatter(X[y_agg == 0, 0], X[y_agg == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(X[y_agg == 1, 0], X[y_agg == 1, 1], s=50, c='green', label='Cluster 2')
plt.scatter(X[y_agg == 2, 0], X[y_agg == 2, 1], s=50, c='blue', label='Cluster 3')
plt.title('Clusters of datapoints')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid()
plt.show()
```
